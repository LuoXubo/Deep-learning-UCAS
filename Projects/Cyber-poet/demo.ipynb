```python
sample = data[:, 1]
print(sample.shape)
tangs = [ix2word[i] for i in sample.numpy()]
print(tangs)
str = ""
for tang in tangs:
    if tang == "<START>":
        continue
    str += tang
    if tang == "。":
        print(str)
        str = ""
```

```python
samples = data[:, :10]
samples = samples.transpose(1, 0)
for sample in samples:
    tangs = [ix2word[i] for i in sample.numpy()]
    str = ""
    for tang in tangs:
        if tang == "<START>" or tang == '</s>':
            continue
        str += tang
        if tang == "。":
            print(str)
            str = ""
        if tang == "<EOP>":
            break
    print("\n\n************\n\n")
```
